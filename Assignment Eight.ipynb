# Import Libraries

```python
import pandas as pd  # To Read The File
import matplotlib.pyplot as plt  # To Draw A Visual Graphs and Charts
```

# Load The File

```python
# Read The File 
path = pd.read_csv('D://smoke_detection_iot.csv')
```

# Calculate The Values Of "Fire Alarm" Column 

```python
val = path['Fire_Alarm'].value_counts()
```

# Draw The Pie Chart

```python
lab = ['Fire', 'No Fire']  # Place The Labels Details
col = ['Red', 'Blue']  # Display Colors
fig, ax = plt.subplots(figsize=(9, 9))  # Control The Size Of The Figure
ax.pie(val, labels=lab, autopct='%1.1f%%', colors=col)  
ax.set_title('Fire Alarm Graph Assignment')
ax.legend(loc="upper right")
plt.show()
```
